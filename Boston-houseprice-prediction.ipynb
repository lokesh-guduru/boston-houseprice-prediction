```python
# Importing basic libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Boston dataset
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
# Separate features and target variable
X = pd.DataFrame(boston.data, columns=boston.feature_names)
y = pd.Series(boston.target, name='PRICE')
```

```python
# Split data into training and testing sets using train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
split_info = {'total_entries': len(X.index), 'train_count': len(X_train.index), 'test_count' : len(X_test.index)}
split_info
```

```python
# Create a Pipeline object to scale the data and train a linear regression model
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
pipeline = Pipeline([('scaler', StandardScaler()), ('regression', LinearRegression())])
pipeline.fit(X_train, y_train)
```

```python
# Predict on the test set and evaluate performance using various metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
reg_pred = pipeline.predict(X_test)
```

```python
## on which parameters the model has been trained
pipeline.get_params()
```

```python
print("Linear Regression Model Metrics:")
print(f"MAE: {mean_absolute_error(y_test, reg_pred)}")
print(f"MSE: {mean_squared_error(y_test, reg_pred)}")
print(f"R2 Score: {r2_score(y_test, reg_pred)}")
```

```python
#Scatter plot for the prediction
sns.regplot(y=reg_pred, x=y_test)
```

```python
residuals = y_test-reg_pred
residuals
```

```python
#Scatterplot WRT prediciton and residuals
#uniform distribution
sns.scatterplot(x=reg_pred, y=residuals)
```

```python
# Create a Pipeline object to scale the data and train a random forest regressor
from sklearn.ensemble import RandomForestRegressor
pipeline = Pipeline([('scaler', StandardScaler()), ('random_forest', RandomForestRegressor())])
random_pred = pipeline.fit(X_train, y_train)
```

```python
# Predict on the test set and evaluate performance using various metrics
random_pred = pipeline.predict(X_test)
print("Random Forest Regressor Metrics:")
print(f"MAE: {mean_absolute_error(y_test, random_pred)}")
print(f"MSE: {mean_squared_error(y_test, random_pred)}")
print(f"R2 Score: {r2_score(y_test, random_pred)}")
```

```python
# Initialize the DataFrame to store the results
results = pd.DataFrame(columns=["Algorithm", "MAE", "MSE", "R2"])
results = results.append({"Algorithm": "Linear Regression Model", "MAE": mean_absolute_error(y_test, reg_pred), "MSE": mean_squared_error(y_test, reg_pred), "R2": r2_score(y_test, reg_pred)}, ignore_index=True)
results = results.append({"Algorithm": "Random Forest Regressor", "MAE": mean_absolute_error(y_test, random_pred), "MSE": mean_squared_error(y_test, random_pred), "R2": r2_score(y_test, random_pred)}, ignore_index=True)
```
